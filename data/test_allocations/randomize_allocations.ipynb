```python
import numpy as np
import pandas as pd

# Define the data
funds = ['voo', 'bsv', 'biv', 'blv', 'vb', 'vo', 'vv', 'vtv', 'vti', 'vwo']
balances = [400000, 100000, 200000, 150000, 50000, 25000, 25000, 25000, 15000, 10000]

# Generate 10 different CSV files
for i in range(1, 11):
    # Generate 10 random numbers that sum to 1 using Dirichlet distribution
    target_allocations = np.random.dirichlet(np.ones(10))
    target_balances = (np.random.dirichlet(np.ones(10)) * 1000000).astype(int)
    
    # Create DataFrame
    df = pd.DataFrame({
        'Index_Fund': funds,
        'Balance': target_balances,
        'Target_Allocation': target_allocations
    })
    
    # Export to CSV using pandas.to_csv()
    filename = f'fund_allocation_{i}.csv'
    df.to_csv(filename, index=False)
    print(f"Created: {filename}")
```
